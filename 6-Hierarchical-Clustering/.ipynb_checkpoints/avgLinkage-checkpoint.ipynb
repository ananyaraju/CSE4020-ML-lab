```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
np.random.seed(0)
cluster1 = np.random.randn(30, 2) + np.array([2, 3])
cluster2 = np.random.randn(30, 2) + np.array([8, 8])
cluster3 = np.random.randn(30, 2) + np.array([5, 12])

data = np.vstack((cluster1, cluster2, cluster3))

plt.scatter(data[:, 0], data[:, 1], s=30)
plt.title('Dataset with Three Clusters')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
def average_linkage_clustering(data):
    clusters = [[point] for point in data]

    while len(clusters) > 1:
        min_avg_distance = float("inf")
        merge_indices = None

        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                cluster1 = clusters[i]
                cluster2 = clusters[j]

                # Calculate the average distance between two clusters
                avg_distance = np.mean([
                    euclidean_distance(point1, point2)
                    for point1 in cluster1
                    for point2 in cluster2
                ])

                if avg_distance < min_avg_distance:
                    min_avg_distance = avg_distance
                    merge_indices = (i, j)

        if merge_indices is not None:
            i, j = merge_indices
            clusters[i].extend(clusters[j])
            del clusters[j]

    return clusters
```

```python
resulting_clusters = average_linkage_clustering(data)
```

```python
Z = linkage(data, method='average')

plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title('Average Linkage Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
