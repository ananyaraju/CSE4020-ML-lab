```python
import numpy as np
import pandas as pd
```

```python
array = np.array([
  [2.7810836,   2.550537003,  0],
  [1.465489372, 2.362125076,  0],
  [3.396561688, 4.400293529,  0],
  [1.38807019,  1.850220317,  0],
  [3.06407232,  3.005305973,  0],
  [7.627531214, 2.759262235,  1],
  [5.332441248, 2.088626775,  1],
  [6.922596716, 1.77106367,   1],
  [8.675418651, -0.242068655, 1],
  [7.673756466, 3.508563011,  1]
]).astype(float)

column_values = ['X1', 'X2', 'Y']

df = pd.DataFrame(data = array, columns = column_values)

df
```

```python
x = df.drop(['Y'], axis = 1)
y = df['Y']
```

```python
coef = [0] * x.shape[1]
inter = 0
predictions = [0] * len(df)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
lr = 0.3
```

```python
epochs = 6
```

```python
for epoch in range(epochs):
    for i in range(len(df)):
        prediction = sigmoid(inter + coef[0]*x.loc[i,"X1"] + coef[1]*x.loc[i,"X2"])
        predictions[i] = prediction   
        inter += lr * (y.loc[i]-prediction) * prediction * (1-prediction) * 1.0
        coef[0] += lr * (y.loc[i]-prediction) * prediction * (1-prediction) * x.loc[i, "X1"]
        coef[1] += lr * (y.loc[i]-prediction) * prediction * (1-prediction) * x.loc[i, "X2"]
    print(df)
#     print(predictions)
    for i, p in enumerate(predictions):
        if (p>0.5):
            predictions[i] = 1
        else:
            predictions[i] = 0
    print(predictions)
    print()
    print()
```
