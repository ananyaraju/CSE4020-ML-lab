```python
import numpy as np
import pandas as pd
import math
```

```python
array = np.array([
    ["sunny","hot","high","weak","no"],
    ["sunny","hot","high","strong","no"],
    ["overcast","hot","high","weak","yes"],
    ["rainy","mild","high","weak","yes"],
    ["rainy","cool","normal","weak","yes"],
    ["rainy","cool","normal","strong","no"],
    ["overcast","cool","normal","strong","yes"],
    ["sunny","mild","high","weak","no"],
    ["sunny","cool","normal","weak","yes"],
    ["rainy","mild","normal","weak","yes"],
    ["sunny","mild","normal","strong","yes"],
    ["overcast","mild","high","strong","yes"],
    ["overcast","hot","normal","weak","yes"],
    ["rainy","mild","high","strong","no"]
])

columns = np.array(["outlook","temperature","humidity","wind","play"])

df = pd.DataFrame(data = array, columns = columns)
df
```

```python
yeses = len(df.loc[df["play"] == "yes"])
nos = len(df.loc[df["play"] == "no"])
print("Y = ", yeses)
print("N = ", nos)
```

```python
def entropy(p, n):
    if (p == 0 or n == 0):
        return 0
    t = p+n
    e = - (n/t)*(math.log2(n/t)) - (p/t)*(math.log2(p/t))
    return round(e, 3)
```

```python
# ENTROPY OF ENTIRE TREE
entropy(yeses, nos)
```

```python
# ENTROPIES BASED ON OUTLOOK

print(df["outlook"].unique())

# SUNNY
df_sunny = df.loc[df["outlook"] == "sunny"]
sunny_yes = len(df_sunny.loc[df["play"] == "yes"])
sunny_no = len(df_sunny.loc[df["play"] == "no"])
n_sunny = sunny_yes + sunny_no
e_sunny = entropy(sunny_yes, sunny_no)

# RAINY
df_rainy = df.loc[df["outlook"] == "rainy"]
rainy_yes = len(df_rainy.loc[df["play"] == "yes"])
rainy_no = len(df_rainy.loc[df["play"] == "no"])
n_rainy = rainy_yes + rainy_no
e_rainy = entropy(rainy_yes, rainy_no)

# OVERCAST
df_overcast = df.loc[df["outlook"] == "overcast"]
overcast_yes = len(df_overcast.loc[df["play"] == "yes"])
overcast_no = len(df_overcast.loc[df["play"] == "no"])
n_overcast = overcast_yes + overcast_no
e_overcast = entropy(overcast_yes, overcast_no)

print("ENTROPIES = ")
print("outlook\t", "yes\t", "no\t", "entropy\t")
print("sunny\t", sunny_yes,'\t', sunny_no,'\t', e_sunny)
print("rainy\t", rainy_yes,'\t', rainy_no,'\t', e_rainy)
print("overcast", overcast_yes,'\t', overcast_no,'\t', e_overcast)

# AVERAGE INFORMATION BASED ON OUTLOOK
total = yeses + nos
ai = e_sunny*n_sunny/total + e_rainy*n_rainy/total + e_overcast*n_overcast/total
ai = round(ai, 3)
print("AVERAGE INFORMATION = ", ai)

# GAIN BASED ON OUTLOOK
gain = entropy(yeses, nos) - ai
print("GAIN = ", gain)
```
