```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)
X = 7 * np.random.rand(100, 1)
y = 2 * X**3 + 8 * X + np.random.rand(100, 1)
```

```python
degree = 4
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
```

```python
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
```

```python
plt.scatter(X, y, color="blue")
plt.plot(X, poly_model.predict(poly_features.fit_transform(X)), color="red")
plt.title("Polynomial Regression with degree = 4")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
y_pred = poly_model.predict(poly_features.fit_transform([[6]]))
print(y_pred)
```
