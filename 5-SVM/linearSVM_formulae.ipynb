```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
positive = [[3,1],[3,-1],[6,1],[6,-1]]
negative = [[1,0],[0,1],[0,-1],[-1,0]]
```

```python
for i in range(4):
    plt.scatter(positive[i][0], positive[i][1], marker='x')
    plt.scatter(negative[i][0], negative[i][1], marker='o')
```

```python
s1 = [negative[0][0], negative[0][1], 1]
s2 = [positive[1][0], positive[1][1], 1]
s3 = [positive[0][0], positive[0][1], 1]
svectors = [s1, s2, s3]
svectors
```

```python
from sympy import symbols, Eq, solve 
  
a1, a2, a3 = symbols('a1,a2,a3') 
   
e1 = Eq((a1 * np.dot(s1, s1) + a2 * np.dot(s1, s2) + a3 * np.dot(s1, s3)), -1) 
print("Equation 1: ", e1)   

e2 = Eq((a1 * np.dot(s2, s1) + a2 * np.dot(s2, s2) + a3 * np.dot(s2, s3)), 1) 
print("Equation 2: ", e2) 

e3 = Eq((a1 * np.dot(s3, s1) + a2 * np.dot(s3, s2) + a3 * np.dot(s3, s3)), 1) 
print("Equation 3: ", e3)  
  
print(solve((e1, e2, e3), (a1, a2, a3))) 
```
