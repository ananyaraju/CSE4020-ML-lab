```python
import numpy as np
import pandas as pd
```

```python
array = np.array([
    [1, 1, 0], 
    [2, 4, 0], 
    [3, 9, 1], 
    [4, 6, 1], 
    [5, 5, 1], 
    [6, 3, 0],              
    [7, 4, 1],
    [8, 9, 1],
    [1, 5, 0], 
    [4, 3, 0],              
    [7, 1, 0],
    [3, 9, 1],
    [0, 0, 0],
    [1, 0, 0],
    [2, 2, 0],
])

columns = ['x','y','sum>10']

df = pd.DataFrame(data = array, columns = columns)
df
```

```python
def euclidean(x1, y1, x2, y2):
    d = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return round(d, 3)
```

```python
maxdist = 0
maxp1 = [0]*2
maxp2 = [0]*2
for i in range(len(df)):
    for j in range(len(df)):
        if i<j:
            dist = euclidean(df.loc[df.index[i], 'x'], df.loc[df.index[i], 'y'], df.loc[df.index[j], 'x'], df.loc[df.index[j], 'y'])
            if (dist > maxdist):
                maxdist = dist
                maxp1 = [df.loc[df.index[i], 'x'], df.loc[df.index[i], 'y']]
                maxp2 = [df.loc[df.index[j], 'x'], df.loc[df.index[j], 'y']]
print(maxdist)
print(maxp1)
print(maxp2)
```

```python
C1 = maxp1
C2 = maxp2
```

```python
for i in range(len(df)):
    df.loc[df.index[i], 'c1_dist'] = euclidean(df.loc[df.index[i], 'x'], df.loc[df.index[i], 'y'], C1[0], C1[1])
    df.loc[df.index[i], 'c2_dist'] = euclidean(df.loc[df.index[i], 'x'], df.loc[df.index[i], 'y'], C2[0], C2[1])
    if df.loc[df.index[i], 'c1_dist'] > df.loc[df.index[i], 'c2_dist']:
        df.loc[df.index[i], 'cluster1'] = 2
    else:
        df.loc[df.index[i], 'cluster1'] = 1
```

```python
newc1 = [0]*2
newc2 = [0]*2
countc1 = 0
countc2 = 0
for i in range(len(df)):
    if df.loc[df.index[i], 'cluster1'] == 1:
        newc1[0] += df.loc[df.index[i], 'x']
        newc1[1] += df.loc[df.index[i], 'y']
        countc1 += 1
    else:
        newc2[0] += df.loc[df.index[i], 'x']
        newc2[1] += df.loc[df.index[i], 'y']
        countc2 += 1
newc1[0] = round(newc1[0]/countc1, 2)
newc1[1] = round(newc1[1]/countc1, 2)
newc2[0] = round(newc2[0]/countc2, 2)
newc2[1] = round(newc2[1]/countc2, 2)
print(newc1)
print(newc2)
```

```python
C1 = newc1
C2 = newc2
```

```python
for i in range(len(df)):
    df.loc[df.index[i], 'c1_dist'] = euclidean(df.loc[df.index[i], 'x'], df.loc[df.index[i], 'y'], C1[0], C1[1])
    df.loc[df.index[i], 'c2_dist'] = euclidean(df.loc[df.index[i], 'x'], df.loc[df.index[i], 'y'], C2[0], C2[1])
    if df.loc[df.index[i], 'c1_dist'] > df.loc[df.index[i], 'c2_dist']:
        df.loc[df.index[i], 'cluster2'] = 2
    else:
        df.loc[df.index[i], 'cluster2'] = 1
```
